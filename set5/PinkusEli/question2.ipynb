```python
import numpy as np
```

```python
Y_train = np.loadtxt('data/train.txt')
Y_test = np.loadtxt('data/test.txt')
# num of users in data
M = max(max(Y_train[:,0]), max(Y_test[:,0])).astype(int)  # users# num of movies in data
N = max(max(Y_train[:,1]), max(Y_test[:,1])).astype(int) # movies
```

```python
Y_train_dic = dict()
for pt in Y_train:
    Y_train_dic[str(int(pt[0])) + ' ' +str(int(pt[1]))] = pt[2]
```

```python
Y_train_dic
```

```python
K = 100
U = np.random.rand(M,K) / 2
V = np.random.rand(N,K) / 2
V.shape
```

```python
U.shape
```

```python
def grad_u (i):
    vecs = []
    for j in range(N):
        key = str(i) + ' ' + str(j)
        if key in Y_train_dic:
            vecs.append(V[j].transpose()*(Y_train_dic[key]-U[i].dot(V[j].transpose())))
    vecs = np.array(vecs)
    return -np.sum(vecs, axis = 0)
```

```python
def grad_v (j):
    vecs = []
    for i in range(M):
        key = str(i) + ' ' + str(j)
        if key in Y_train_dic:
            vecs.append(U[i].transpose()*(Y_train_dic[key] - U[i].dot(V[j].transpose())))
    vecs = np.array(vecs)
    result =  -np.sum(vecs, axis = 0)
    return result
```

```python
EPOCHS = 50
eta = 0.03
for epoc in range(EPOCHS):
    print(epoc)
    i_perm = np.random.permutation(range(M))
    for i in i_perm:
        gradu = grad_u(i)
        print(type(gradu))
        U[i] -= eta * grad_u(i).transpose()
        
    j_perm = np.random.permutation(range(N))
    for j in j_perm:
        gradv = grad_v(j)
        print(type(gradv))
        V[j] -= eta * grad_v(j).transpose()
```
