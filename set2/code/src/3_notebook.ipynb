```python
import numpy as np
from matplotlib import pyplot as plt
from numpy import genfromtxt
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
%matplotlib inline
```

## Load data

```python
train_file = 'data/problem3data.txt'
train_data = genfromtxt(train_file, delimiter='\t')

y_train = train_data[:, 9]
X_train = train_data[:, :9]

alphas = np.linspace(0.01,3,30)
```

```python
# get ws for the alpha values
ws = []
for alpha in alphas:
    clf = Lasso(alpha = alpha)
    clf.fit(X_train,y_train)
    ws.append(clf.coef_)
    
```

```python
# flip the order of the array to help with ploting
flip_ws = [[pt[i] for pt in ws] for i in range(len(ws[0]))]
for i in range(len(flip_ws)):
    plt.plot(alphas, flip_ws[i], label = 'i = ' + str(i+1))
plt.xlabel('alpha')
plt.ylabel('$w_i$')
plt.legend(loc = 'upper right')
plt.title('w values vs. regularization strength for Lasso')
```

```python
# get ws for the alpha values
alphas_ridge = list(range(1,10001))
ws_ridge = []
for alpha in alphas_ridge:
    clf = Ridge(alpha = alpha)
    clf.fit(X_train,y_train)
    ws_ridge.append(clf.coef_)
```

```python
# flip arrays for ploting
flip_ws_ridge = [[pt[i] for pt in ws_ridge] for i in range(len(ws_ridge[0]))]
for i in range(len(flip_ws_ridge)):
    plt.plot(alphas_ridge, flip_ws_ridge[i], label = 'i = ' + str(i+1))
plt.xlabel('alpha')
plt.ylabel('$w_i$')
plt.legend(loc = 'upper right')
plt.title('w values vs. regularization strength for Ridge')
```
