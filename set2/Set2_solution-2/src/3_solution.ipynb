```python
import numpy as np
from matplotlib import pyplot as plt
from numpy import genfromtxt
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
%matplotlib inline
```

```python
def train_ridge_model(Lambdae, X_train, y_train):
    coefs = []
    for Lambda in Lambdae:
        clf = Ridge(alpha=Lambda)
        clf.fit(X_train, y_train)
        coefs.append(clf.coef_)
    return coefs


def train_lasso_model(Lambdae, X_train, y_train):
    coefs = []
    for Lambda in Lambdae:
        clf = Lasso(alpha=Lambda)
        clf.fit(X_train, y_train)
        coefs.append(clf.coef_)
    return coefs
```

```python
ridge_lambdae = range(1, 10001)
lasso_lambdae = np.linspace(0.01, 3, 30)

train_file = '../data/problem3data.txt'
train_data = genfromtxt(train_file, delimiter='\t')

y_train = train_data[:, 9]
X_train = train_data[:, :9]

ridge_coefs = train_ridge_model(ridge_lambdae, X_train, y_train)
lasso_coefs = train_lasso_model(lasso_lambdae, X_train, y_train)
```

```python
for i in range(len(ridge_coefs[0])):
    plt.plot(
        ridge_lambdae, [x[i] for x in ridge_coefs],
        label='$w_{i}$'.format(i=str(i + 1)))
plt.legend(loc=1)
plt.xlabel('$\lambda$')
plt.ylabel('$w$')
plt.title('Ridge Regression')
plt.show()

for i in range(len(lasso_coefs[0])):
    plt.plot(
        lasso_lambdae, [x[i] for x in lasso_coefs],
        label='$w_{i}$'.format(i=str(i + 1)))
plt.legend(loc=1)
plt.xlabel('$\lambda$')
plt.ylabel('$w$')
plt.title('Lasso Regression')
plt.show()
```
